Разработать программу на языке высокого уровня, реализующую обход графов в ширину и глубину. Для хранения вершин следует использовать стек или очередь из стандартной библиотеки языка. Граф должен быть представлен в виде списка смежности. Результат обхода (последовательность вершин) выводится на экран

```python
import graphviz
import csv
from queue import Queue
```

```python
def stdcsvread(FILENAME):
    adj = {}
    with open(FILENAME) as file:
        reader = csv.reader(file)
        for row in reader:
            st = int(row[0])
            adj[st] = []
            for a in row[1:]:
                adj[st].append(int(a))
    return adj
```

```python
def grviz(adj):
    graph = graphviz.Digraph() 
    for a in adj.keys():
        for b in adj[a]:
            graph.edge(str(a),str(b))
    return graph
```

```python
def BFS(adj,st):
    X = {}
    for a in adj.keys():
        X[a] = 0
    T = Queue()
    v_ = st
    T.put(v_)
    X[v_] = 1
    while 0 in X.values() and T!=[]:
        v = T.get()
        print (v,)
        for v__ in adj[v]:
            if X[v__] == 0:
                X[v__] = 1
                T.put(v__)
    while not T.empty():
            print(T.get())
    if( 0 in X.values()):
        print("Граф обойти не удалось")
```

```python
def DFS(adj, st):
    X = {}
    for a in adj.keys():
        X[a] = 0
    T =[]
    v_ = st
    T.append(v_)
    X[v_] = 1
    while 0 in X.values() and T!=[]:
        v = T.pop()
        print (v,)
        for v__ in adj[v]:
            if X[v__] == 0:
                X[v__] = 1
                T.append(v__)
    while T!=[]:
            print(T.pop())
    if( 0 in X.values()):
        print("Граф обойти не удалось")
```
